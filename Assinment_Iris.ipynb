```python
import numpy as np

# Introducing missing values
df.iloc[0, 0] = np.nan  # Introducing a missing value in the first row of 'sepal_length'
df.iloc[2, 2] = np.nan  # Introducing a missing value in the third row of 'petal_length'
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Fill missing values only in numeric columns
df_numeric = df.select_dtypes(include=[np.number])
df[df_numeric.columns] = df_numeric.apply(lambda x: x.fillna(x.mean()), axis=0)
df.head()
```

```python
additional_df = pd.DataFrame({
    'species': ['setosa', 'versicolor', 'virginica'],
    'sepal_ratio': [0.7, 0.9, 0.85]
})
```

```python
merged_df = pd.merge(df, additional_df, on='species')
merged_df.head()
```

```python
# Create a pivot table
pivot_table = df.pivot_table(values='sepal_length', index='species', columns='petal_length', aggfunc='mean')
pivot_table
```
